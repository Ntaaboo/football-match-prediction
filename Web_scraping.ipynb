```python
#fetching content from a url using the requests library
import requests
```

```python
#defining the url
standings_url = 'https://fbref.com/en/comps/11/2023-2024/2023-2024-Serie-A-Stats'
```

```python
#downloading the page , using the get method
data = requests.get(standings_url)
```

```python
#viewing the html
data.text
```

```python
#parsing html with beautiful soup
from bs4 import BeautifulSoup
```

```python
#Initialized the soup object
soup = BeautifulSoup(data.text)
```

```python
#selecting the table using css selector, selecting all table elements
standings_table = soup.select('table.stats_table')[0]
```

```python
#finding all anchor tags in our table
links = standings_table.find_all('a')
#get href property of each link
links = [l.get("href") for l in links]
#filtering links to only get squad links
links = [l for l in links if '/squads/' in l]
links
```

```python
#converting links into full urls (retrieving the absolute links)
team_urls = [f"https://fbref.com{l}" for l in links]
```
